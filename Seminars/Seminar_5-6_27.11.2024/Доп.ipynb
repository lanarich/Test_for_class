```python
import math
```

```python
def apply_func(x, func):
    y = []
    for i in range(0, len(x)):
        y.append(func(x[i]))
    return y
```

```python
y = [1,5,9,10]
x = [0,1,2,3]

plt.plot(y, marker='^', ms=7, color='green', lw=2, linestyle='--')
```

```python
x = list(range(0, 20))
y = apply_func(x, math.sqrt)
```

```python
x2 = list(range(0, 3))
y2 = apply_func(x2, math.exp)
```

```python
plt.plot(y, ms=3, color='green', lw=2, label='График 1')
plt.plot(y2, ms=3, color='blue', lw=1, label='График 2')
plt.legend(fontsize=10, loc='lower left')
```
