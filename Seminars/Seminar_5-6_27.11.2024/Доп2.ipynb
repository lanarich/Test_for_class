### *Exercise 2.3* (Advanced)

Write a code that will replace missings in the dataset according to the following rules:

- if a type of a column is NOT "object": replace missings with rounded mean of the column 
- if a type of a column is "object": replace missings with mode of the column 

**Hints**

- Use a "for" loop to iterate over df columns 
- Use if/else condition to distinguish between column types (google for the details)
- Use .dtype to check the type of a column 
- Use "round" function to get rounded mean (google for the details)
- Use .mean() and .mode() functions to get mean and mode. 
- Use df.isna().sum() to check the results 

```python
#Print your code here

columns = list(df)
for i in columns:
    if df[i].dtype != "object":
        df[i] = df[i].fillna(round(df[i].mean(), 0))
    else: 
        df[i] = df[i].fillna(df[i].mode()[0])
        
df.isna().sum()
```
